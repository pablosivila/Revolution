# Vamos limpar a coluna Case Number (já tratada anteriormente) e quebrá-la em ano, mês e dia.

```python
import re
```

```python
df['Year2'] = df['Case Number'].apply(lambda x: int(re.findall('([0-9]{4}).[0-9]{2}.[0-9]{2}', x)[0]))
df['month'] = df['Case Number'].apply(lambda x: int(re.findall('[0-9]{4}.([0-9]{2}).[0-9]{2}', x)[0]))
df['day'] = df['Case Number'].apply(lambda x: int(re.findall('[0-9]{4}.[0-9]{2}.([0-9]{2})', x)[0]))
```

```python
df.info()
```

```python
df['day'].value_counts()
```

```python
# Atenção especial para 721 ocorrências de dia 0 e uma ocorrência de dia 38. Entenderemos dia 0 como não sendo especificado.
# e dia 38 é claramente um erro de input.
```

```python
df[df['day']==0]
```

```python
#Como agora temos todas as informações da coluna 'Case Number' devidamente organizada em colunas 
# de ano, mês e dia, podemos retirá-la de nosso dataset.Seguindo a mesma lógica, retiraremos também a coluna
# 'Date', que além de oferecer informação redundante ainda a oferece em formato irregular.
```

```python
df.drop(columns=['Case Number'], inplace =True)
```

```python
df.drop(columns=['Date'], inplace =True)
```

```python
# Vamos verificar discrepâncias entre a coluna recém criada ('Year2')  e a coluna 'Year' que tínhamos anteriormente:
```

```python
df[df['Year2']!=df['Year']]
```
